```python
#1. random
import random
print(random.random())  #print only decimal no. 0 or 1 ke bich me sirf
```

```python
# randint
import random
print(random.randint(1,10))  #random number generate  krega 1,10 ke bich me or int value he dega 
```

```python
# randrange
import random
print(random.randrange(1,10)) # 1-9 tak ki range but 10 included nhi
```

```python
#Choice
import random
num=1,2,3,4,5
print(random.choice(num)) # series me se koi b random number de dega
```

```python
#choices
import random

lst = [1,2,3]
print(random.choices(lst, k=2))

tup=(1,2,3,4,5)
print(random.choices(tup, k=3))

```

```python
#6. shuffle
import random
lst = [1,2,3,4]
random.shuffle(lst)
print(lst) # list ko shuffle karta he har bar list ka order change ho jata he 
```

```python
#7.sample
import random

lst = [1,2,3,4]
print(random.sample(lst, 2))

name="rahul"
print(random.sample(name,3))  # random characters de dega string me se


```

```python
#uniform
import random
num,num2=1,10
print(random.uniform(num,num2))   # do integer value ke bich me float(decimal) number generate karta he random or value 1,or 10 b de skta he 
```

```python
#1. ceil
import math
num=5.5
print(math.ceil(num)) #ceil() ka matlab hota hai number ko upar wale nearest integer tak round karna  value float me he honi chahiye
```

```python
#2. floor -----floor() ka matlab hota hai number ko niche wale nearest integer tak round karna 7.9 ki lowest or nearest value batayega
import math
num=7.9
print(math.floor(num))
```

```python
#3. sqrt
import math
num=4
print(math.sqrt(num))  #  sqr. root nikalta he number ka 
```

```python
#4.pow
import math
num=2
print(math.pow(num,2)) #Power (exponent) nikalta hai
```

```python
#5.factorial
import math
fact=4
print(math.factorial(fact))  #number ka factorial nikalega only int.
```

```python
#6.fabs 
import math
num=-10
print(math.fabs(num))  #Value (hamesha positive)  
```

```python
#7.gcd gratest common division
import math
num1,num2=10,15
print(math.gcd(num1,num2))
```

```python
#8. log-----------------Ye logarithm base e (≈ 2.718) nikalta hai
import math
num=10
print(math.log(num))    #e ko kis power pe uthaye ki 10 mile”
```

```python
#9.log10 --------------------Ye base 10 ka logarithm nikalta hai
import math
num=100
print(math.log10(num))    #“10 ko kis power pe uthaye ki num mile”
```

```python
#10.exp ------------------Ye e^x (e power x) nikalta hai
import math
num=2
print(math.exp(num))
```

# ### **Datetime Module**

1. datetime
2. date

4. timedelta

6. today

```python
#1.date time
import datetime
from datetime import datetime
print(datetime.now())  # current date + time deta hai
```

```python
#2.date
from datetime import date
print(date.today())#current date dega 
```

```python
#3.time
from datetime import time
now=datetime.now()
print(now.time())#sirf current time nikalta he
```

```python
#4.timedelta--------------timedelta date ko add ya subtract karne ke kaam aata hai
from datetime import timedelta
print(timedelta(days=6))

from datetime import datetime, timedelta
today=datetime.now()
ft=today+timedelta(days=5) #aaj se 5 din baad ki date
print(ft)



```

```python
#5.now
from datetime import datetime
print(datetime.now())  #current date & time
```

```python
#6 today------------current date  deta he 
from datetime import date
print(datetime.today())
```

```python
#7.strftime------------------Date ko string format me convert karta hai
from datetime import datetime
now = datetime.now()
print(now.strftime("%d-%m-%Y"))
print(now.strftime("%d-%B-%Y %H:%M:%S"))
```

```python
#8.strptime---------string ko date me convert kar diya
from datetime import datetime
d = datetime.strptime("26-03-2026", "%d-%m-%Y")
print(d)
```

```python
#9.timestamp------------Date/time ko seconds (Unix format) me convert karta hai
from datetime import datetime
now=datetime.now()
print(now.timestamp())
```

```python
#10.fromtimestamp
from datetime import datetime
print(datetime.fromtimestamp(8000000))
```
