```python
#. Convert 25 into binary and print it.
num=25
print("The decimal number 25 into binary is: ",bin(num))
```

```python
#Convert binary 0b1010 into decimal.
num=0b1010
print("Binary 0b1010 into Decimal: ",int(num))
```

```python
# Add two numbers and print the result in binary.
num1=4
num2=3
addition=num1+num2
print(" The result in binary is: ",bin(addition))
```

```python
# Convert int → binary → back to int and check type.
num=10
#int into binary
binary=bin(num)
print("Int into Binary:",binary)
print("Type:",type(binary))

integer=int(binary,2)
print("Binary into int:",integer)
print("Type:",type(integer))
```
