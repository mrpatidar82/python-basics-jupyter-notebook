```python
# Type-casting :

#Type casting is the process in which one data type is converted into another data type.


#---int to => float,bool,complex,string typecasting---------------------------------------------
age=21
print(age,type(age))# integer

#int to float
b=float(age)
print(b,type(b))

#int to boolean
c=bool(age)
print(c,type(c))

#int to complex
d=complex(age)
print(d,type(d))

# int to string 
e=str(age)
print(e,type(e))
```

```python
#string to => float,bool,complex,int typecasting
name="Rahul"
print(name,type(name))#this is a string


b=bool(name)
print(b,type(b))#str to bool 

c=complex(name)
print(c,type(c)) # str to complex -----Error---complex() arg is a malformed string

d=int(name)
print(d,type(d))  #str to int -----Error---invalid literal for int() with base 10: 'Rahul'

e=float(name)
print(e,type(e)) # str to float ---------Error--------could not convert string to float: 'Rahul'
```

```python
#Float to => string,bool,complex,int typecasting
a=25.5
print(a,type(a))  # float

b=str(a)
print(b,type(b)) #float to string

c=bool(a)
print(c,type(c)) #float to bool

d=complex(a)
print(d,type(d))# float to complex

e=int(a)
print(e,type(e)) # float to int
```

```python
#Bool to => string,float,complex,int typecasting.
a=True
print(a,type(a)) # bool

b=str(a)
print(b,type(b))#bool to string

c=float(a)
print(c,type(c))#bool to float

d=complex(a)
print(d,type(d))#bool to complex

e=int(a)
print(e,type(e))# bool to integer
```
